### Parsing BLAST results

In computer programming parsing is the process of analysing a file according to a set of rules that are specified in the code.  To parse a file you need to know something about the structure of the text in the file.  For example in the above BLAST results example each new result begins with the line BLASTP 2.2.26 [Sep-21-2011].  Thus you could determine the number of separate results by counting the times this line appears.  The fasta header from each query sequence also appears in each result starting with the line Query=.  

```python
#!/usr/bin/env python

# Example 11.1 
# Name: blast_results_count.py
# Description:  This program counts the number of separate blast results in a file
# Usage: python blast_results_count.py filename 
# Example: python blast_results_count.py NC_010001_results.txt

import sys

# Add a check to see if there is the right number of arguments 
# Exit if there is not the correct number
if len(sys.argv) != 2 :
	print "This program takes a filename as an argumnet"
	sys.exit(2)
    
# Open the file
blast_file = open(sys.argv[1], 'r')

# read the file into the list blast_results
blast_results = blast_file.readlines()

# Initialize counts
count = 0

for line in blast_results:
	if line.startswith("BLASTP 2.2.26"): 
		count += 1
print "There are %s separate query results in this file." % (count)
```

### Setting flags

One way to capture the just the first blast hit would be to use flags to determine if this is a new result and change the flag once the top hit has been read. 

```python
#!/usr/bin/env python

# Example 11.2 
# Name: flag_top_hit.py
# Description:  This program captures the sequenc alignment for the top blast hit
# Usage: python flag_top_hit.py filename 
# Example: python flag_top_hit.py NC_010001_5_results2.txt

import sys

# Add a check to see if there is the right number of arguments 
# Exit if there is not the correct number
if len(sys.argv) != 2 :
	print "This program takes a filename as an argumnet"
	sys.exit(2)
    
# Open the file
blast_file = open(sys.argv[1], 'r')

# read the file into the list blast_results
blast_results = blast_file.readlines()

new_blast_flag = 'FALSE'

for line in blast_results:
	if line.startswith("Sequences producing significant alignments"):
		new_blast_flag = 'TRUE'
	if (new_blast_flag == 'TRUE') and line.startswith("ref|"): 
		print line
		new_blast_flag = 'FALSE'
```

Flags are also set using 1 and 0 instead of TRUE and FALSE. Below is an example that captures all the top BLAST hits.

```python
#!/usr/bin/env python

# Example 11.2 
# Name: flag_hit_list.py
# Description:  This program captures the sequenc alignment for the top blast hit
# Usage: python flag_hit_list.py filename 
# Example: python flag_hit_list.py NC_010001_5_results2.txt

import sys

# Add a check to see if there is the right number of arguments 
# Exit if there is not the correct number
if len(sys.argv) != 2 :
	print "This program takes a filename as an argumnet"
	sys.exit(2)
    
# Open the file
blast_file = open(sys.argv[1], 'r')

# read the file into the list blast_results
blast_results = blast_file.readlines()

new_blast_flag = 'FALSE'

for line in blast_results:
	if line.startswith("Sequences producing significant alignments"):
		new_blast_flag = 'TRUE'
		print "NEW BLAST RESULT"
	if (new_blast_flag == 'TRUE') and line.startswith("ref|"): 
		print line
	if (new_blast_flag == 'TRUE') and line.startswith(">"):  
		new_blast_flag = 'FALSE'
```
